```python
print('best out of what?')
out_of = int(input())
print("two player or one player?(type 2 or 1)")
mode = int(input())-1
cur_amount = 0
ties = 0
wins = 0
loses = 0
```

```python
class HOW_DID_THE_AI_LOSE(Exception): pass
def display(x, o):
    print(' y: 0 1 2')
    print('    _ _ _')
    for i in range(3):
        a = []
        for j in range(3):
            if [i, j] in x:
                a.append("x")
            elif [i, j] in o:
                a.append("o")
            else:
                a.append(" ")
        print('x:'+str(i)+'|'+a[0]+"|"+a[1]+"|"+a[2]+'|')

        print('    _ _ _')

def checkwin(board):
    for i in range(3):
        work1 = True
        work2 = True
        for j in range(3):
            if not [i, j] in board:
                work1 = False
            if not [j, i] in board:
                work2 = False
        if work1 or work2:
            return True
    work1 = True
    work2 = True
    for i in range(3):
        if not [i, i] in board:
            work1 = False
        if not [i, 2-i] in board:
            work2 = False
    return work1 or work2

def twop():
    x = []
    o = []
    move = 1
    exit = False
    while not checkwin(x) and not checkwin(o):
        if len(x)+len(o) == 9:
          exit = True
        display(x, o)
        move+=1
        move %= 2
        if move == 0:
            print("x move")
        else:
            print("o move")
        work = False
        while not work:
            try:
                a = [int(i) for i in input().split() if int(i)<=2]
                if len(a)<2 or a in x or a in o:
                    raise Exception("among")
                work = True
                if move == 0:
                    x.append(a)
                else:
                    o.append(a)
            except KeyboardInterrupt:
                exit = True
                break
            except:
                print("please input a space seperated pair of integers from 0 to 2")
        if exit:
            break
    if not exit: display(x, o)
    if checkwin(x):
        print("x wins")
    if checkwin(o):
        print("o wins")

    else:
        print("among")
def onep():
    import random
    global cur_amount
    global wins
    global loses
    global ties
    p = []
    b = []
    move = 1
    possible = []
    exit = False
    for i in range(3):
        possible+=[[i, j] for j in range(3)]
    while not checkwin(p) and not checkwin(b):
        move+=1
        move%=2
        if len(possible) == 0:
            break
        if move == 0:
            display(p, b)
            print("your turn(type in the format (x y))")
            work = False
            while not work:
                try:
                    test = input()
                    if test == "exit":
                        exit = True
                        break
                    a = [int(i) for i in test.split() if int(i)<=2]
                    if len(a)<2 or (not a in possible):
                        raise Exception("among")
                    work = True
                    p.append(a)
                    possible.remove(a)
                except KeyboardInterrupt:
                    exit = True
                    break
                except:
                    print("please input a space seperated pair of integers from 0 to 2 that has not been taken yet")
        else:
            #display(p, b)
            moved = False
            for i in possible:
                if checkwin(b+[i]):
                    moved = True
                    b.append(i)
                    possible.remove(i)
                if moved:
                    break
            if not moved:
                for i in possible:
                    if checkwin(p+[i]):
                        moved = True
                        b.append(i)
                        possible.remove(i)
                    if moved:
                        break

            if not moved:
                i = random.choice(possible)
                b.append(i)
                possible.remove(i)
        if exit:
            break
    if not exit: display(p, b)
    if checkwin(p):
        print("you win")
        wins += 1
    elif checkwin(b):
        print("you lose")
        loses+=1
    else:
        print("Ties")
    cur_amount +=1
if mode:
  twop()
else:
  onep()
if cur_amount == out_of:
  try:
    print(str(100*(1-(out_of/loses)))+'% were ties/wins')
  except:
    print('100% were ties/wins')
if not mode:
  try:
    if 1-(out_of/loses) > 0.50:
      raise HOW_DID_THE_AI_LOSE("the ai got so mad because it lost the tic tac toe and exploded and died and yeah")
  except:
    raise HOW_DID_THE_AI_LOSE("the ai got so mad because it lost the tic tac toe and exploded and died and yeah")
```
